```python

data = torch.tensor(digits.data,dtype=torch.float32)
data = DataLoader(data)
```

```python
device = torch.device("cpu")
```

```python

class NeuralNet(nn.Module):
    def __init__(self):
    
        super(NeuralNet,self).__init__()
        self.relu = nn.ReLU()
        self.enc = self.encoder()
        self.dec = self.decoder()
    
    
    def encoder(self):
    
        return nn.Linear(8,8,bias=False)
    
    def decoder(self):
    
        return nn.Linear(8,8,bias=False)
    
    def forward(self,x):
        
        
        
        out = self.enc(x)
        out = self.relu(out)
        out = self.enc(out)
    
        out = self.dec(out)
        out = self.relu(out)
        out = self.dec(out)
        
        return out
```

```python
# training
model = NeuralNet()
creatrion = nn.L1Loss()

optim = torch.optim.Adam(model.parameters(),lr=0.0001)
```

```python
l1_loss = []


for i in range(10):
    for tx in range(digits.data.shape[0]):
        image = next(iter(data)).reshape(8,8)
        
        outputs = model(image.reshape(8,8))
        
        loss = creatrion(outputs,image)
        
        optim.zero_grad()
        loss.backward()
        
        optim.step()
        
        if i%150 == 0:
            
        
            l1_loss.append(loss.item())
            print(f"Loss is {loss.item()}")

            
```
