```python
import numpy as np
from new import NewWorld
from collections import defaultdict
```

```python
env = NewWorld()
```

```python
n_states = 4*4
n_actions = 4
```

```python
xy2i = lambda x,y:x*env.size+y
```

```python
def greedy_policy(Q,state,eps=0.2):
    
    def policy_fn(observation):
    
        A = np.ones(n_actions, dtype=float) * eps / n_actions
        best_action = np.argmax(Q[observation])
        A[best_action] += (1.0 - eps)
        return A
    return policy_fn
```

```python
def sarsa(eps = 0.1,step_size=0.5):
    
    Q = defaultdict(lambda:np.zeros(n_actions))
    
    discount = 1.0
    
    for i in range(1): # episode
        
        s = 0
        
        env.done = False
        policy = greedy_policy(Q,s,eps)
        
        action = np.argmax(policy(s))
        
        while True:
            
            n_s,reward,done,_ = env.step(action)
            
            n_s = xy2i(n_s[0],n_s[1])
            
            print("state =>",s,"next state =>",n_s,"action =>",action,"done =>",done)
            
            n_a = np.argmax(policy(n_s))
          
            Q[s][action] = Q[s][action] + step_size * (reward + discount * (Q[n_s][n_a] - Q[s][action]))

            s = n_s
            action = n_a
            
            print(Q)
            
            if done == True:
                break
    
    return Q

    
    
```

```python
output = sarsa()
```
