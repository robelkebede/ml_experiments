```python
import numpy as np
from new import NewWorld
import matplotlib.pyplot as plt
import time
```

```python
env = NewWorld()

env.state
```

```python
n_states = 4*4
```

```python
xy2i = lambda x,y:x*env.size+y
discount_factor = 1.0

def lookahead(state,value):
        
        A = np.zeros(4)
        for a in range(4):
            for prob, next_state, reward, done in env.P[state][a]:
                A[a] += prob * (reward + discount_factor * value[next_state])
                
        return A

```

```python
value = np.zeros(16) # create this in the new file 

theta=0.0001 #theta
i=0

while True:
    
    delta = 0 # delta
    
    for s in range(n_states):
        
        A = lookahead(s,value)
        best_action = np.max(A)
        
        delta = max(delta, np.abs(best_action - value[s])) # optimal

        value[s] = best_action # update
   
        print(delta)
    
    if i == 1000:
        break

    if delta < theta:
        break

    i+=1
    
policy = np.zeros([n_states, 4])
for s in range(n_states):
    # One step lookahead to find the best action for this state
    A = lookahead(s,value)
    best_action = np.argmax(A)
    # Always take the best action
    policy[s, best_action] = 1.0
```
