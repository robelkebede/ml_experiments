```python
import numpy as np
from new import NewWorld
from collections import defaultdict
```

```python
env = NewWorld()
```

```python
n_actions = 4
n_states = 4*4
```

```python
def greedy_policy(Q,eps,n_actions):
    
    
    def policy(obs):
        
        A = np.ones(n_actions) * eps / n_actions
        
        best_action = np.argmax(Q[obs])
        
        A[best_action] += (1.0 - eps)
        
        return A
    return policy
        
```

```python
def q_max(Q,a,state,n_state):
    
    m = []

    for i in range(4):
        
        val = Q[n_state][i] - Q[state][a]

        m.append(m)
    
    return np.argmax(m)
        
            
    
    
```

```python
xy2i = lambda x,y:x*env.size+y
```

```python
def q_learning_off_policy(step_size = 0.5,eps= 0.1,discount = 1.0):
    
    Q = defaultdict(lambda: np.zeros(n_actions))

    
    for i in range(1):
        
        env.reset()
        s = 0
        policy = greedy_policy(Q,eps,n_actions)
        env.done = False

        
        while True:
            
            
            action = np.argmax(policy(s))
            
            n_s,reward,done,_ = env.step(action)
            
            n_s = xy2i(n_s[0],n_s[1])
            
            print("state =>",s,"next state =>",n_s,"action =>",action,"done =>",done)
            
            val = np.max([Q[n_s][i] - Q[s][action] for i in range(4)])
            
            Q[s][action] = Q[s][action] + step_size * (reward + discount * val)
            
            s = n_s
            
            
            print(Q)
            
            if done == True:
                break
            
```
